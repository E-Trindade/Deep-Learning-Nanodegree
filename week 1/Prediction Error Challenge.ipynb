```python
%matplotlib inline
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
#import seaborn
import numpy as np
```

```python
dataframe = pd.read_csv('data/challenge_dataset.txt', sep=',', names=['x', 'y'])
dataframe
```

# Building Linear Regression Model

```python
lr_body = linear_model.LinearRegression()
x_values = dataframe[['x']]
y_values = dataframe[['y']]
lr_body.fit(x_values, y_values)
```

## Angular Coeficient

```python
lr_body.coef_
```

## Y Coeficient

```python
lr_body.intercept_
```

```python
plt.scatter(x_values, y_values)
plt.plot(x_values, lr_body.predict(x_values))
plt.show()
```

```python
dataframe['predicted'] = lr_body.predict(x_values)
dataframe['error'] = np.absolute(dataframe['predicted'] - dataframe['y'])
dataframe
```

```python
actual = dataframe['y']
predicted = dataframe['predicted']

fig, ax = plt.subplots()
ax.bar(actual.index, actual, width=0.5)
ax.bar(predicted.index  + 0.5, predicted, width=0.5)

fig.show()
```
